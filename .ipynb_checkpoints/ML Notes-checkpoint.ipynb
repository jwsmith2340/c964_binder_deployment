```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

from IPython.display import display

pd.set_option('display.max_columns', 40)
```

```python
def build_cus_df(cus_age, cus_gender, cus_dependent, cus_education, cus_marital, cus_income, cus_card, cus_time, 
                 cus_inactive, cus_contact_ct, cus_cred_limit, cus_revolve, cus_open, cus_perc_chg_q4_q2, 
                 cus_count_chg_q4_q2, cus_trans_amt, cus_trans_ct, cus_util):
    cus_df = pd.DataFrame({'Customer_Age': [cus_age],
                           'Gender': [cus_gender],
                           'Dependent_count': [cus_dependent],
                           'Education_Level': [cus_education],
                           'Marital_Status': [cus_marital],
                           'Income_Category': [cus_income],
                           'Card_Category': [cus_card],
                           'Months_on_book': [cus_time],
                           'Credit_Limit': [cus_cred_limit],
                           'Total_Revolving_Bal': [cus_revolve],
                           'Avg_Open_To_Buy': [cus_open],
                           'Total_Amt_Chng_Q4_Q1': [cus_perc_chg_q4_q2],
                           'Total_Ct_Chng_Q4_Q1': [cus_count_chg_q4_q2],
                           'Total_Trans_Amt': [cus_trans_amt],
                           'Total_Trans_Ct': [cus_trans_ct],
                           'Avg_Utilization_Ratio': [cus_util]
                          })

    return cus_df
```

```python
# Import the data set
ds = pd.read_csv('BankChurners.csv')
ds.rename(columns = {'Attrition_Flag':'Existing_Customer_Bit'}, inplace = True)
ds.head()
```

```python
# Drop unused/unnecessary columns
ds.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
         'CLIENTNUM',
         'Total_Relationship_Count',
         'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], 
        inplace=True, 
        axis=1)
```

```python
ds[:15]
```

```python
# Turning String categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Split data into Features and Labels
X = ds.drop('Existing_Customer_Bit', axis=1)
y = ds['Existing_Customer_Bit']

# Anything that is a category, included customer bit which is going to be my target value, 
# so check here if there are issues running the ML model
categorical_features = ['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)],
                               remainder = 'passthrough')
transformed_X = transformer.fit_transform(X)
transformed_X
pd.DataFrame(transformed_X[:15])
```

```python
ds.loc[ds['Existing_Customer_Bit'] == 'Existing Customer', 'Existing_Customer_Bit'] = 1
ds.loc[ds['Existing_Customer_Bit'] == 'Attrited Customer', 'Existing_Customer_Bit'] = 0
```

```python
# Split data into Features and Labels
X = transformed_X

#pd.DataFrame(X)
```

```python
y.head()
```

```python
# Split the data into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Despite a 37% accuracy score, this effectively demonstrates the one_hot 
# functionality to encode string type values to num values
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
# Both work, but Random Forest Classifier is more accurate
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression

y = ds['Existing_Customer_Bit']

X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train, y_train.astype(int))

clf.score(X_test, y_test.astype(int))
```

## Making predictions with the Model

```python
# Manually comparing the predcictions to the truth labels to evaluate the model
y_preds = clf.predict(X_test)
np.mean(y_preds == y_test)
```
