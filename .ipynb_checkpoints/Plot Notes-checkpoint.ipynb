## Matplot Lib

## sklearn

```python
# Testing dropping all columns with non-number values
ds.drop(['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category', 'Total_Amt_Chng_Q4_Q1', 'Total_Ct_Chng_Q4_Q1'], 
        inplace=True, 
        axis=1)
```

```python
ds.loc[ds['Existing_Customer_Bit'] == 'Existing Customer', 'Existing_Customer_Bit'] = 1
ds.loc[ds['Existing_Customer_Bit'] == 'Attrited Customer', 'Existing_Customer_Bit'] = 0

# Make dataset2 only the non-existing customers
ds = ds.loc[ds['Existing_Customer_Bit'] == 0]

# Changed from customer_bit to age
X = ds.drop('Customer_Age', axis=1)
y = ds['Customer_Age']

ds.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train);
```

```python
X_train
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
